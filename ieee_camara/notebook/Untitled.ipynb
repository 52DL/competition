```python
import numpy as np
import glob
import csv
from tqdm import tqdm
from conditional import conditional
from os.path import join
import pandas as pd
from sklearn.utils.extmath import softmax
CLASSES = [
    'HTC-1-M7',
    'iPhone-6',     
    'Motorola-Droid-Maxx',
    'Motorola-X',
    'Samsung-Galaxy-S4',
    'iPhone-4s',
    'LG-Nexus-5x', 
    'Motorola-Nexus-6',
    'Samsung-Galaxy-Note3',
    'Sony-NEX-7']
TEST_FOLDER = '../input/test/'
pd.set_option('display.height', 10000)
pd.set_option('display.max_rows', 5000)
pd.set_option('display.max_columns', 5000)
pd.set_option('display.width', 10000)
```

```python
des = np.load('../subm/densenet201_98_0.05239213012534667_0.030665695160758077_0.9842336166712367_0.8787061994609164_tta_arithmetic.npy')
dpn = np.load('../subm/dpn131_68_0.22852461410608763_0.04059238999003643_0.9812174389909515_0.9376731301939059_tta_arithmetic.npy')
ir = np.load('../subm/inceptionresnetv2_115_0.2800604541126328_0.022912003500223682_0.9757334795722512_0.9158686730506156_tta_arithmetic.npy')
res = np.load('../subm/resnet152_154_0.21108618408131102_0.03966159357182812_0.972717301891966_0.9236016371077762_tta_arithmetic.npy')
nxt = np.load('../subm/resnext101_64x4d_106_0.05270318477823023_0.021929375704770025_0.9839594187003016_0.9042253521126761_tta_arithmetic.npy')
```

```python
log_des,log_dpn,log_ir,log_res,log_nxt = softmax(des),softmax(dpn),softmax(ir),softmax(res),softmax(nxt)
```

```python
#mean = (log_des+log_dpn+log_ir+log_res+log_nxt)/5
mean = (des+res+nxt)/3
```

```python
ids = glob.glob(join(TEST_FOLDER, '*.tif'))
ids.sort()
with conditional(True, open('submision_log_3mean.csv', 'w')) as csvfile:
    csv_writer = csv.writer(csvfile, delimiter=',',quotechar='|', quoting=csv.QUOTE_MINIMAL)
    csv_writer.writerow(['fname', 'camera'])
    for i, idx in enumerate(tqdm(ids)):
        prediction_class_idx = np.argmax(mean[i])
        csv_writer.writerow([idx.split('/')[-1], CLASSES[prediction_class_idx]])
        
```
